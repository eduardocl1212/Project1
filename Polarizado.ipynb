```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
funda = cv2.imread('/content/drive/My Drive/DICIS/5.jpeg')
plt.imshow(funda)
plt.show()
```

```
hsv_funda = cv2.cvtColor(funda, cv2.COLOR_RGB2HSV)
light_orange = (0,100,100)
dark_orange = (10,255,255)
mask = cv2.inRange(hsv_funda, light_orange, dark_orange)
result = cv2.bitwise_and(funda, funda, mask=mask)
plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```
